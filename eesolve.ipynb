```python
from sage.interfaces.singular import singular

# 1. Define parameters as the base ring (Include N_const here)
ParamRing.<alpha_1, alpha_2, beta_1, beta_2, gamma_1, gamma_2, iota, N_0> = PolynomialRing(QQ)
K = ParamRing.fraction_field()

# 2. Define state variables over K using 'lex' order
R_ring.<S, L_1, L_2, H_1, H_2> = PolynomialRing(K, order='degrevlex')

D = alpha_1 + alpha_2

# 3. Define the original polynomial system
dSdt = - beta_1*S*H_1 - beta_2*S*H_2 + iota*(N_0 - S - L_1 - L_2 - H_1 - H_2)
dL_1dt = (beta_1*S*H_1 - alpha_1*L_1)*D - alpha_2*beta_2*L_1*H_2
dL_2dt = (beta_2*S*H_2 - alpha_2*L_2)*D - alpha_1*beta_1*L_2*H_1
dH_1dt = (alpha_1*L_1 - gamma_1*H_1)*D + alpha_1*beta_1*L_2*H_1
dH_2dt = (alpha_2*L_2 - gamma_2*H_2)*D + alpha_2*beta_2*L_1*H_2
# dRdt is omitted because it is structurally redundant

# 4. Define the integrated conservation equation
#conservation_eq = S + L_1 + L_2 + H_1 + H_2 + R_var - N_0

# 5. Build the ideal with exactly 6 independent constraints
I = ideal([dSdt, dL_1dt, dL_2dt, dH_1dt, dH_2dt]) #, conservation_eq])

print(f"Ideal dimension: {I.dimension()}") # This will now output 

gb = I.groebner_basis()

print("Made degrevlex")
#print(gb)

R_lex.<S, L_1, L_2, H_1, H_2> = PolynomialRing(QQ, order='lex')

phi = R_ring.hom(R_lex.gens(), R_lex)

# 3. Map your degrevlex basis into the lex ring via the homomorphism
gb_lex_mapped = [phi(poly) for poly in gb]

# 4. Construct the ideal in the lex ring and run FGLM safely
I_lex = R_lex.ideal(gb_lex_mapped)
gb_lex = I_lex.groebner_basis(algorithm='fglm')

print(gb_lex)

# # 4. Convert to lex using the FGLM algorithm
# gb_lex = I.change_ring(R_lex).groebner_basis(algorithm='fglm')

# print(gb_lex)
```
